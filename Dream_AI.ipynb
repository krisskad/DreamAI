# **Enter Prompt**

```python
prompt = "kids toddler toys"
style_id = 78 # Realistic v2 Style
```

# **Generate Image**

```python
data = {
    "is_premium": False,
    "input_spec": {
        "prompt": prompt,
        "style": style_id,
        "display_freq": 10
    }
}

gen_response = requests.post('https://paint.api.wombo.ai/api/v2/tasks', headers=headers, data=json.dumps(data))

try:
  image_id = gen_response.json()["id"]
  while True:
    response = requests.get(f'https://paint.api.wombo.ai/api/v2/tasks/{image_id}', headers=headers)
    time.sleep(3)
    try:
      img = response.json()["result"]["final"]
      urllib.request.urlretrieve(img, f"{image_id}.jpg")
      break
    except:
      continue
except Exception as e:
  print(f'ERROR : {e}')
  print(gen_response.text)
```
